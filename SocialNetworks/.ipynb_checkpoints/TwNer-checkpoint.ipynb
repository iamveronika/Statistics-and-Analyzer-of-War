```python
import os
import csv
import gzip
import spacy
import numpy as np
import pandas as pd
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
path0='0819_UkraineCombinedTweetsDeduped.csv.gzip'
data0 = pd.read_csv(path0, compression='gzip')
display(data0[0:5])
```

```python
texts=data0['acctdesc'].dropna().tolist()
print(texts[0:3])
```

```python
texts2=[]
for text in tqdm(texts[0:10000]):
    doc = nlp(text)
    for ent in doc.ents:
        texts2+=[[ent.text, ent.label_,1]]
print(texts2[0:10])
```

```python
data2=pd.DataFrame(data=texts2,columns=['name','label','count'])
data3=data2.groupby(['name','label'],as_index=False).sum()
display(data3)
```

```python
cols=data3['label'].unique().tolist()
print(cols)
```

```python
data=data3[data3['label']=='PERSON'].sort_values('count',ascending=False)[0:20]
display(data)
```

```python
plt.figure(figsize=(15,6))
plt.title('Tweet Ranking on PERSON')
g = sns.barplot(x=data['name'],y=data['count'],color="#3F0FB7",data=data)
g.tick_params(axis='x', labelrotation=90)
plt.show()
```
